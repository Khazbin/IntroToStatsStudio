## Section 2: Distribution of our data:

```python
# Plot the distribution of your variable using distplot
sb.distplot(heartDF['anaemia'])
plt.show()

sb.distplot(heartDF['platelets'])
plt.show()
```

```python
# Create boxplot to show distribution of variable
fig, ax = plt.subplots()
ax.boxplot(heartDF['anaemia'], vert=False, showmeans=True, meanline=True)
plt.show()

fig, bx = plt.subplots()
bx.boxplot(heartDF['platelets'], vert=False, showmeans=True, meanline=True)
plt.show()
```

```python
# Feel free to add any additional graphs that help you answer the questions below.
```

```python
# Another way to check the skewness of our variable
skewAnaemia = heartDF['anaemia'].skew()
skewPlatelets = heartDF['platelets'].skew()
print(f"Anaemia: ", skewAnaemia,"\nPlatelets: ",skewPlatelets)
```

```python
# Another way to check the kurtosis of our variable
kurtAnaemia = heartDF['anaemia'].kurtosis()
kurtPlatelets = heartDF['platelets'].kurtosis()
print(f"Anaemia: ",kurtAnaemia,"\nPlatelets: ",kurtPlatelets)
```

### Interpretation of how our data is distributed by variable?
<b>Looking at the above graphs, what can you tell about the distribution of your variables?</b>
<br>
Anaemia has two groups, which makes perfect sense, with 0 having more points than 1. Platelets has a bit of clustering on the low end but then spreads out rather far toward the right.<br><br>
<b>What is the skewness and kurtosis of your variables.  What does this mean?</b>
<br>
The skewness of Anaemia is 0.28 which means there is a right skew. The skewness of Platelets is 1.46 implying a greater right skew than Anaemia.<br>
The kurtosis of Anaemia is -1.94, which means this data is flatter than a normal distribution. The kurtosis of Platelets is positive and over 6, indicating that the data is tightly peaked.<br><br><br>
<b>What are some of the differences you note looking at a categorical variable vs a continuous variable?</b>
<br>
Categorical variables are discrete with specific possibilities for what they can be (i.e. only integers, 0 or 1, specific strings) and are counted rather than summed. In this set, most categorical data will have two peaks when graphed and no outliers. They won't have any notable skewing and the kurtosis will be flat. Continuous variables exist within a range and can be mathematically manipulated. In this set, the continuous data will have variable skewing and kurtosis and may have outliers.

## Section 3: Finding Correlations

Lets start by breaking our data into two.  

```python
# splitting the dataframe into 2 parts
# on basis of ‘DEATH_EVENT’ column values
diedDF = heartDF[heartDF['DEATH_EVENT'] == 1 ]
livedDF = heartDF[heartDF['DEATH_EVENT'] == 0 ]
```

```python
# Plot your variable based on if they died or lived

sb.distplot(diedDF['platelets'])
sb.distplot(livedDF['platelets'])
plt.title("Chances of survival vs Variable")
plt.legend(('Died','Lived'))
plt.plot()
```

```python
# Feel free to add any additional graphs that help you answer the questions below.

sb.distplot(diedDF['anaemia'])
sb.distplot(livedDF['anaemia'])
plt.title("Chances of survival vs Variable")
plt.legend(('Died','Lived'))
plt.plot()
```
